```python
%run "galg-computations.ipynb"

# Examples from the paper "3-braid knots with maximal 4-genus" by Baader, Lewark, Misev, Truöl 
braids = []
braids.append([1,1,1,2,2,2,1,1,2,2,1,1,2,2])
braids.append([1,2,1,1,1,1,2,2,1,1,2,2,1,1])
braids.append([1,2,1,1,1,1,1,1,2,2,1,1,2,2,1,1])
braids.append([1,2,1,1,1,1,2,2,2,1,1,2,2,2,1,1])
braids.append([1,2,1,1,2,1,1,2,1,1,1,1,2,2,1,1])

SeifertMatrices = []
for b in braids:
    SeifertMatrices.append(Link(BraidGroup(3)(b)).seifert_matrix())

# Continue searching until g - g_alg >= 2 is proven
for A in SeifertMatrices:
    defect = 0
    while defect < 2:
        defect, T = galg(A,1,1)
        if defect > 1:
            print(str(list(T)))
```
